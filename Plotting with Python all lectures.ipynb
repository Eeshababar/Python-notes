```python
mass = 10
if mass > 10:
    print('Mass is large')
else:
    print('Mass is small')
```

```python

Mass = [10,20,30,40,60,70]
for m in Mass:
    if m > 30:
        print('Mass is large')
    else:
        print('Mass is small')
        
```

```python
Mass = [1,2,3,4,5,10,20,30,40,60,70]
for m in Mass:
    if m > 40:
        print('Mass is large')
    elif m > 10:
        print('Mass is fine')
    else:
        print('Mass is negligible')
```

```python
i = 0
mass = [10,20,30,40,50,60,70]
Acceleration = [1,2,3,4,5,6,7]

for i in range(7):
    if (mass[i] > 40) and (Acceleration[i] > 4):
        print('High force will be produced')
    elif (mass[i] > 20) and (Acceleration[i] > 2):
         print('Enough force will be produced')
    else:
        print('Unsufficient force will be produced')
    
```

```python
i = 0
mass = [10,20,30,40,50,60,70]
Acceleration = [1,2,3,4,5,6,7]

for i in range(7):
    if (mass[i] > 40) and (Acceleration[i] > 4):
        print('High force will be produced')
    elif (mass[i] > 20) and (Acceleration[i] > 2):
         mass[i] = mass[i]+20
         print('Enough force will be produced', mass[i])
    else:
        print('Unsufficient force will be produced')
```

```python

Force  = 10.0
for i in range(5): # execute the loop 5 times
    print(i, ':', Force)
    if Force > 20.0:
        print('Much high force')
        Force = Force - 5.0
    else:
        print('Need more force')
        Force = Force + 10.0
print('final Force:', Force)
```

# Functions

```python
def print_date(Year,month,day):
    date = str(Year) + '/' + str(month) + '/' + str(day)  
    print(date)
```

```python
print_date(21,11,5)
```

```python
def average(numbers):
    if len(numbers) == 0:
        return None
    return sum(numbers) / len(numbers)
```

```python
a = average([1,2,3,4])
print(a)
```

```python
print(average([-1,2,-3,4]))
```

# Scope of variable

```python
pressure = 103.9


def adjust(t):
    temp = t * 1.43 / pressure
    return temp
```

```python
print('adjusted:', adjust(0.9))
print('Pressure value used is:', pressure) #Pressure is global variable
print('temperature after call:', temp) #temp is local variable
```
